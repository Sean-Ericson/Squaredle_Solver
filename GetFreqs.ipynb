```python
import requests
import time
import pickle
from tqdm.notebook import trange
```

```python
_wait = 0.5

def get_freq(term):
    response = None
    while True:
        try:
            response = requests.get('https://api.datamuse.com/words?sp='+term+'&md=f&max=1').json()
        except:
            print('Could not get response. Sleep and retry...')
            time.sleep(_wait)
            continue
        break
    freq = 0.0 if len(response)==0 else float(response[0]['tags'][0][2:]) / int(1e6)
    return freq
```

```python
with open("words.txt", 'r') as f:
    words = [line.strip() for line in f.readlines()]
```

```python
freqs = dict()
with open("freqs.pkl", 'rb') as f:
    freqs = pickle.load(f)
for i in trange(len(words)):
    word = words[i]
    if word in freqs.keys():
        continue
    freqs[word] = get_freq(word)
    with open("freqs.pkl", 'wb') as f:
        pickle.dump(freqs, f)
```

```python
def get_word_freq_dict():
    freqs = {}
    with open("freqs.pkl", "rb") as f:
        freqs = pickle.load(f)
    return freqs

freqs = get_word_freq_dict()
```

```python
for word,freq in freqs.items():
    if freq == 0:
        if get_freq(word) != 0:
            print(word, freq)
```
